```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'

df = pd.read_csv('../data/processed/processed_faers.csv')
print(df.shape)
df.head()
```

```python
print("=== 컬럼 목록 ===")
print(df.columns.tolist())
print("\n=== 결측치 ===")
print(df.isnull().sum())
print("\n=== 데이터 타입 ===")
print(df.dtypes)
```

```python
top_reac = df['pt'].value_counts().head(20)

plt.figure(figsize=(12, 6))
top_reac.plot(kind='bar', color='steelblue')
plt.title('Top 20 Adverse Reactions')
plt.xlabel('Reaction')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_drug = df['drugname'].value_counts().head(20)

plt.figure(figsize=(12, 6))
top_drug.plot(kind='bar', color='tomato')
plt.title('Top 20 Drugs by Report Count')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sex_counts = df['sex'].value_counts()

plt.figure(figsize=(6, 6))
sex_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#4e79a7', '#f28e2b', '#bab0ac'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```
